## Preprocessing

- Numerical Columns: Columns like sBytesMax, rBytesMax, sBytesMin, etc., have some missing values but not a very high percentage.
- Categorical Columns: Columns like sIPs and rIPs have missing values but again, not a very high percentage.

Given the analysis, we decide to fill in the missing values to avoid losing potentially useful data. For numerical columns, we will use the median, and for categorical columns, we will use the mode.

To train the classification model, we need to transform categorical data into numeric for consistency and to apply algorithms RF and LSTM. We used `LabelEncoder` for this transformation. Additionally, we applied a scaler to preprocess the data, ensuring that numerical features are on a similar scale, which helps improve the performance of LSTM.

```python
numerical_cols = westermo_df.select_dtypes(include=['number']).columns
westermo_df[numerical_cols] = westermo_df[numerical_cols].fillna(westermo_df[numerical_cols].median())

categorical_cols = westermo_df.select_dtypes(include=['object']).columns
print(f"Categorial columns:\n{categorical_cols}")
for col in categorical_cols:
    westermo_df[col] = westermo_df[col].fillna(westermo_df[col].mode()[0])

# print(westermo_df.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
category_mappings = {}

for col in categorical_cols:
    le = LabelEncoder()
    westermo_df[col] = le.fit_transform(westermo_df[col])
    label_encoders[col] = le
    category_mappings[col] = {str(original): int(encoded) for original, encoded in zip(le.classes_, le.transform(le.classes_))}
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
westermo_df[numerical_cols] = scaler.fit_transform(westermo_df[numerical_cols])
```

```python
from imblearn.over_sampling import SMOTE # type: ignore
from imblearn.combine import SMOTETomek

# Handling Class Imbalance
X = westermo_df.drop(['IT_M_Label', 'IT_B_Label', 'NST_M_Label', 'NST_B_Label'], axis=1)
y = westermo_df['IT_M_Label']

smote = SMOTE()
X_res, y_res = smote.fit_resample(X, y)

# smote_tomek = SMOTETomek(random_state=42)
# X_res, y_res = smote_tomek.fit_resample(X, y)
```

## Feature Selection

Using RF classifier from `scikit-learn`

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
rf_classifier = RandomForestClassifier(
    n_estimators=100, 
    random_state=42, 
    n_jobs=-1, 
    class_weight='balanced'
)
rf_classifier.fit(X_train, y_train)
y_pred = rf_classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.model_selection import cross_val_score

conf_matrix = confusion_matrix(y_test, y_pred)
acc_score = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
cv_scores = cross_val_score(rf_classifier, X_train, y_train, cv=5)

print("Confusion Matrix:\n", conf_matrix)
print("Accuracy Score:", acc_score)
print("Classification Report:\n", class_report)
print("Cross-validation Scores:", cv_scores)
```

```python
import numpy as np

# Feature Importance
feature_importances = rf_classifier.feature_importances_
features = X.columns
indices = np.argsort(feature_importances)[::-1]
```

```python
import matplotlib.pyplot as plt

# Plot Feature Importance
plt.figure(figsize=(10, 6))
plt.title("Feature Importance")
plt.bar(range(X.shape[1]), feature_importances[indices], align="center")
plt.xticks(range(X.shape[1]), features[indices], rotation=90)
plt.show()
```

```python
# Feature Selection
top_features = features[indices][:10]  # Top 10
print(top_features)
X_train_selected = X_train[top_features]
X_test_selected = X_test[top_features]
```
