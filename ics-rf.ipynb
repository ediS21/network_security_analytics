## Preprocessing

- Numerical Columns: Columns like sBytesMax, rBytesMax, sBytesMin, etc., have some missing values but not a very high percentage.
- Categorical Columns: Columns like sIPs and rIPs have missing values but again, not a very high percentage.

Given the analysis, we decide to fill in the missing values to avoid losing potentially useful data. For numerical columns, we will use the median, and for categorical columns, we will use the mode.

To train the classification model, we need to transform categorical data into numeric for consistency and to apply algorithms RF and LSTM. We used `LabelEncoder` for this transformation. Additionally, we applied a scaler to preprocess the data, ensuring that numerical features are on a similar scale, which helps improve the performance of LSTM.

```python
numerical_cols = ics_df.select_dtypes(include=['number']).columns
ics_df[numerical_cols] = ics_df[numerical_cols].fillna(ics_df[numerical_cols].median())

categorical_cols = ics_df.select_dtypes(include=['object']).columns
print(f"Categorial columns:\n{categorical_cols}")
for col in categorical_cols:
    ics_df[col] = ics_df[col].fillna(ics_df[col].mode()[0])

# print(ics_data.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
category_mappings = {}

for col in categorical_cols:
    le = LabelEncoder()
    ics_df[col] = le.fit_transform(ics_df[col])
    label_encoders[col] = le
    category_mappings[col] = {str(original): int(encoded) for original, encoded in zip(le.classes_, le.transform(le.classes_))}
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
ics_df[numerical_cols] = scaler.fit_transform(ics_df[numerical_cols])
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTETomek

# Handling Class Imbalance
X = ics_df.drop(['IT_M_Label', 'IT_B_Label', 'NST_M_Label', 'NST_B_Label'], axis=1)
y = ics_df['IT_M_Label']

smote = SMOTE()
X_res, y_res = smote.fit_resample(X, y)

# smote_tomek = SMOTETomek(random_state=42)
# X_res, y_res = smote_tomek.fit_resample(X, y)
```

## Feature Selection

Using RF classifier from `scikit-learn`

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
rf_classifier = RandomForestClassifier(
    n_estimators=100, 
    random_state=42, 
    n_jobs=-1, 
    class_weight='balanced'
)
rf_classifier.fit(X_train, y_train)
y_pred = rf_classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.model_selection import cross_val_score

conf_matrix = confusion_matrix(y_test, y_pred)
acc_score = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
cv_scores = cross_val_score(rf_classifier, X_train, y_train, cv=5)

print("Confusion Matrix:\n", conf_matrix)
print("Accuracy Score:", acc_score)
print("Classification Report:\n", class_report)
print("Cross-validation Scores:", cv_scores)
```

```python
import numpy as np

# Feature Importance
feature_importances = rf_classifier.feature_importances_
features = X.columns
indices = np.argsort(feature_importances)[::-1]
```

```python
import matplotlib.pyplot as plt

# Plot Feature Importance
plt.figure(figsize=(10, 6))
plt.title("Feature Importance")
plt.bar(range(X.shape[1]), feature_importances[indices], align="center")
plt.xticks(range(X.shape[1]), features[indices], rotation=90)
plt.show()
```

```python
# Feature Selection
top_features = features[indices][:10]  # Top 10
print(top_features)
X_train_selected = X_train[top_features]
X_test_selected = X_test[top_features]
```

### RF classifier from `cuml`

```python
from cuml.ensemble import RandomForestClassifier as cuRF
import numpy as np

# Ensure data is in float32 format for pickling
X_train_selected = X_train_selected.astype(np.float32)
y_res = y_res.astype(np.float32)

rf_cuml = cuRF(n_estimators=100, max_depth=20, random_state=42, n_streams=1)
rf_cuml.fit(X_train_selected, y_train)
y_pred_cuml = rf_cuml.predict(X_test_selected)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Evaluation
conf_matrix = confusion_matrix(y_test, y_pred_cuml)
acc_score = accuracy_score(y_test, y_pred_cuml)
class_report = classification_report(y_test, y_pred_cuml)

print("Confusion Matrix:\n", conf_matrix)
print("Accuracy Score:", acc_score)
print("Classification Report:\n", class_report)
```

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = []

for train_index, val_index in kf.split(X_train):
    X_train_fold, X_val_fold = X_train.iloc[train_index], X_train.iloc[val_index]
    y_train_fold, y_val_fold = y_train.iloc[train_index], y_train.iloc[val_index]
    
    rf_cuml.fit(X_train_fold, y_train_fold)
    y_val_pred = rf_cuml.predict(X_val_fold)
    fold_acc_score = accuracy_score(y_val_fold, y_val_pred)
    cv_scores.append(fold_acc_score)

print("Cross-validation Scores:", cv_scores)
print("Mean Cross-validation Score:", np.mean(cv_scores))
```

## Improvements and Finetuning

- Total Possible Combinations = 4 * 4 * 3 * 3 * 3 * 3 = 576
- Number of Iterations: 10

`RandomizedSearchCV` will evaluate 10 randomly chosen combinations out of the 576 possible combinations

### From model loading

```python
import pandas as pd
import joblib
import numpy as np
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report

# Load dataset
ics_df = pd.read_csv('ics-flow/preprocessed_ics.csv')

# Feature and target selection
X = ics_df.drop(['IT_M_Label', 'IT_B_Label', 'NST_M_Label', 'NST_B_Label'], axis=1)
y = ics_df['IT_M_Label']

# Handle class imbalance with SMOTE
smote = SMOTE()
X_res, y_res = smote.fit_resample(X, y)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42, shuffle=True)

# Select top features
top_features = ['endDate', 'startOffset', 'end', 'endOffset', 'startDate', 'start',
                'sAckDelayAvg', 'rInterPacketAvg', 'sPayloadSum', 'rBytesSum']
X_train_selected = X_train[top_features].astype(np.float32)
X_test_selected = X_test[top_features].astype(np.float32)

y_res = y_res.astype(np.float32)

# Define parameter grid for RandomizedSearchCV
param_grid = {
    "n_estimators": [100, 200, 300, 400],
    'max_depth': [10, 20, 30, 40],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False],
    "criterion": ["gini", "entropy"],
}

# Load pre-trained RandomForest model
rf_model = joblib.load('ics_rf_cuml_model.pkl')
```

```python
# Perform RandomizedSearchCV
random_search_rf = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=param_grid,
    n_iter=5,
    scoring='accuracy',
    cv=3,               # 3-fold cross-validation
    random_state=42,
    n_jobs=-1,          # Utilize all cores
    verbose=2
)

# Fit the model
random_search_rf.fit(X_train_selected, y_train)

# Output the best parameters and score
print("Best Parameters:", random_search_rf.best_params_)
print("Best Score:", random_search_rf.best_score_)
```

```python
# # Save the best model
# joblib.dump(random_search_rf.best_estimator_, 'best_rf_model.pkl')

# # Predict using the best model
y_pred = random_search_rf.best_estimator_.predict(X_test_selected)
print(classification_report(y_test, y_pred))
```
